```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = yf.download("AAPL",period="1y",interval="1d")['Close']
```

```python
data
```

```python
plt.plot(data.index,data)
plt.xlabel("Date")
plt.ylabel("Closing price")
plt.show()
```

```python
The stock price of apple showed fluctuations throughout the last year. But the graph can be divided into three major regions
1. Jan-March: The stock price observed an overall downward trend in this period (or the trend was bearish roughly).
2. March-May: Although there were ups and downs, the stock price observed an overall sideways trend in this period.
3. May-Dec: The prices fluctuated but showed a upward trend in this period (or the trend was bullish roughly).
```
